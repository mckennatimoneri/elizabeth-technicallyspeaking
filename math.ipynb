```python
import sys
import subprocess
import importlib.util
from pathlib import Path

# Install missing packages into the current Jupyter kernel if needed.
required_packages = ["numpy", "gensim", "nltk", "scipy"]
for package in required_packages:
    if importlib.util.find_spec(package) is None:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])

import numpy as np
from gensim import corpora, models
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from scipy.special import rel_entr
import nltk

# Ensure required NLTK resources are available.
nltk.download("punkt_tab", quiet=True)
nltk.download("punkt", quiet=True)
nltk.download("stopwords", quiet=True)

# ---------------------------------------------------------
# 1. Load prompts and preprocess
# ---------------------------------------------------------

def load_prompts(path):
    prompts = {}
    current_key = None
    lines = []
    with open(path, "r", encoding="utf-8") as f:
        for raw in f:
            line = raw.rstrip("\n")
            if line.startswith("===") and line.endswith("==="):
                if current_key is not None:
                    prompts[current_key] = "\n".join(lines).strip()
                current_key = line.strip("=").strip()
                lines = []
            else:
                lines.append(line)
        if current_key is not None:
            prompts[current_key] = "\n".join(lines).strip()
    return prompts

prompt_path = Path("article2_prompts.txt")
if not prompt_path.exists():
    raise FileNotFoundError(f"Could not find {prompt_path}. Make sure your notebook is running from the repository root.")

bor_prompts = load_prompts(prompt_path)

stop_words = set(stopwords.words("english"))

def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [t for t in tokens if t.isalpha() and t not in stop_words]
    return tokens

# Use the two selected documents from the prompt file.
doc1 = bor_prompts["OG"]
doc2 = bor_prompts["P4_Perp"]

tokens1 = preprocess(doc1)
tokens2 = preprocess(doc2)

# ---------------------------------------------------------
# 2. Build dictionary + corpus
# ---------------------------------------------------------
texts = [tokens1, tokens2]
dictionary = corpora.Dictionary(texts)
corpus = [dictionary.doc2bow(t) for t in texts]

# ---------------------------------------------------------
# 3. Train LDA model
# ---------------------------------------------------------
num_topics = 5
lda = models.LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    passes=20,
    random_state=42
)

# ---------------------------------------------------------
# 4. Get topic distributions for each document
# ---------------------------------------------------------
def get_topic_vector(lda_model, bow, num_topics):
    """Return a dense topic probability vector."""
    topic_dist = lda_model.get_document_topics(bow, minimum_probability=0)
    vec = np.zeros(num_topics, dtype=float)
    for topic_id, prob in topic_dist:
        vec[topic_id] = prob
    return vec

vec1 = get_topic_vector(lda, corpus[0], num_topics)
vec2 = get_topic_vector(lda, corpus[1], num_topics)

# ---------------------------------------------------------
# 5. KL Divergence
# ---------------------------------------------------------
def kl_divergence(p, q, eps=1e-12):
    p = np.asarray(p, dtype=np.float64) + eps
    q = np.asarray(q, dtype=np.float64) + eps
    p /= p.sum()
    q /= q.sum()
    return np.sum(rel_entr(p, q))

kl_12 = kl_divergence(vec1, vec2)
kl_21 = kl_divergence(vec2, vec1)

print("KL(doc1 || doc2):", kl_12)
print("KL(doc2 || doc1):", kl_21)
```

```python
import sys
import subprocess
import importlib.util
from pathlib import Path

# ---------------------------------------------------------
# Robustly install spaCy if missing
# ---------------------------------------------------------
def ensure_package_installed(package_name):
    if importlib.util.find_spec(package_name) is None:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package_name])

ensure_package_installed("spacy")

# Load spaCy model (download if missing)
import spacy
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    subprocess.check_call([sys.executable, "-m", "spacy", "download", "en_core_web_sm"])
    nlp = spacy.load("en_core_web_sm")

from collections import Counter
import numpy as np

# ---------------------------------------------------------
# Load prompts from file (FIXED NEWLINE BUG)
# ---------------------------------------------------------
def load_prompts(path):
    prompts = {}
    current_key = None
    lines = []
    with open(path, "r", encoding="utf-8") as f:
        for raw in f:
            line = raw.rstrip("\n")   # FIXED
            if line.startswith("===") and line.endswith("==="):
                if current_key is not None:
                    prompts[current_key] = "\n".join(lines).strip()   # FIXED
                current_key = line.strip("=").strip()
                lines = []
            else:
                lines.append(line)
        if current_key is not None:
            prompts[current_key] = "\n".join(lines).strip()   # FIXED
    return prompts

prompt_path = Path("preamble_prompts.txt")
if not prompt_path.exists():
    raise FileNotFoundError(f"Could not find {prompt_path}. Run this notebook from the repository root.")

preamble_prompts = load_prompts(prompt_path)

# ---------------------------------------------------------
# POS Distribution Function
# ---------------------------------------------------------
def pos_distribution(text):
    doc = nlp(text)
    tags = [token.pos_ for token in doc]
    counts = Counter(tags)
    total = sum(counts.values())
    return {tag: count / total for tag, count in counts.items()}

# POS Drift (difference between OG and another prompt)
def pos_drift(dist1, dist2):
    all_tags = set(dist1.keys()) | set(dist2.keys())
    return {tag: abs(dist1.get(tag, 0) - dist2.get(tag, 0)) for tag in all_tags}

# ---------------------------------------------------------
# Detect all keys from preamble_prompts file
# ---------------------------------------------------------
all_keys = list(preamble_prompts.keys())
print("Detected keys:", all_keys)

# Identify OG
if "OG" not in all_keys:
    raise KeyError("Could not find 'OG' in preamble_prompts. Check your prompt labels.")

original_key = "OG"
comparison_keys = [k for k in all_keys if k != original_key]

# ---------------------------------------------------------
# Compute POS distributions for OG + all prompts
# ---------------------------------------------------------
pos_results = {}
pos_results[original_key] = pos_distribution(preamble_prompts[original_key])

for key in comparison_keys:
    pos_results[key] = pos_distribution(preamble_prompts[key])

# ---------------------------------------------------------
# Compute POS drift for each prompt vs OG
# ---------------------------------------------------------
pos_drifts = {}
og_dist = pos_results[original_key]

for key in comparison_keys:
    pos_drifts[key] = pos_drift(og_dist, pos_results[key])

# ---------------------------------------------------------
# Print results
# ---------------------------------------------------------
print("\n=== POS DISTRIBUTIONS ===")
for key, dist in pos_results.items():
    print(f"\n{key}:")
    for tag, prob in sorted(dist.items()):
        print(f"  {tag}: {prob:.4f}")

print("\n=== POS DRIFT (vs OG) ===")
for key, drift in pos_drifts.items():
    print(f"\nOG vs {key}:")
    for tag, diff in sorted(drift.items()):
        if diff > 0:
            print(f"  {tag}: {diff:.4f}")
```

```python
import sys
import subprocess
import importlib.util
from pathlib import Path

# Ensure spaCy + zss installed
def ensure_package_installed(package_name):
    if importlib.util.find_spec(package_name) is None:
        subprocess.check_call([sys.executable, "-m", "pip", "install", package_name])

ensure_package_installed("spacy")
ensure_package_installed("zss")

# Load spaCy model
import spacy
try:
    nlp = spacy.load("en_core_web_sm")
except OSError:
    subprocess.check_call([sys.executable, "-m", "spacy", "download", "en_core_web_sm"])
    nlp = spacy.load("en_core_web_sm")

from zss import Node, simple_distance

# ---------------------------------------------------------
# Load prompts from file
# ---------------------------------------------------------
def load_prompts(path):
    prompts = {}
    current_key = None
    lines = []
    with open(path, "r", encoding="utf-8") as f:
        for raw in f:
            line = raw.rstrip("\n")
            if line.startswith("===") and line.endswith("==="):
                if current_key is not None:
                    prompts[current_key] = "\n".join(lines).strip()
                current_key = line.strip("=").strip()
                lines = []
            else:
                lines.append(line)
        if current_key is not None:
            prompts[current_key] = "\n".join(lines).strip()
    return prompts

prompt_path = Path("article2_prompts.txt")
article2_prompts = load_prompts(prompt_path)

# ---------------------------------------------------------
# Convert spaCy dependency tree в†’ ZSS tree
# ---------------------------------------------------------
def to_tree(token):
    node = Node(token.dep_)
    for child in token.children:
        node.addkid(to_tree(child))
    return node

def sentence_tree_edit_distance(sent1, sent2):
    doc1 = nlp(sent1)
    doc2 = nlp(sent2)

    root1 = [t for t in doc1 if t.head == t][0]
    root2 = [t for t in doc2 if t.head == t][0]

    tree1 = to_tree(root1)
    tree2 = to_tree(root2)

    return simple_distance(tree1, tree2)

# ---------------------------------------------------------
# Compute TED for OG vs all prompts
# ---------------------------------------------------------
all_keys = list(article2_prompts.keys())
original_key = "OG"
comparison_keys = [k for k in all_keys if k != original_key]

ted_results = {}

for key in comparison_keys:
    ted_results[key] = sentence_tree_edit_distance(
        article2_prompts[original_key],
        article2_prompts[key]
    )

# ---------------------------------------------------------
# Print results
# ---------------------------------------------------------
print("\n=== TREE EDIT DISTANCE (vs OG) ===")
for key, dist in ted_results.items():
    print(f"{key}: {dist}")
```
