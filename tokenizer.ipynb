```python
import tiktoken

# Use a model tokenizer compatible with most OpenAI text models.
# If the model-specific tokenizer is unavailable, fall back to cl100k_base.
def get_tokenizer(model_name: str = "gpt-4o-mini"):
    try:
        return tiktoken.encoding_for_model(model_name)
    except Exception:
        return tiktoken.get_encoding("cl100k_base")


def tokenize_text(text: str, model_name: str = "gpt-4o-mini") -> list[int]:
    encoder = get_tokenizer(model_name)
    return encoder.encode(text)


def count_tokens(text: str, model_name: str = "gpt-4o-mini") -> int:
    return len(tokenize_text(text, model_name))


def display_tokenization(text: str, model_name: str = "gpt-4o-mini"):
    encoder = get_tokenizer(model_name)
    tokens = encoder.encode(text)
    print(f"Model: {model_name}")
    print(f"Text: {text!r}")
    print(f"Token count: {len(tokens)}")
    print(f"Tokens: {tokens}")
    print(f"Decoded text: {encoder.decode(tokens)}")
```

```python
text = input("Enter text to tokenize: ")
if not text.strip():
    raise ValueError("Please enter some text to tokenize.")
display_tokenization(text)
```

```python
# Example usage without user input.
sample_text = "Hello! Tokenize this sentence and count the tokens."
display_tokenization(sample_text)
```
