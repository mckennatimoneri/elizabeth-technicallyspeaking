```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# --- CONFIGURATION ---
# The order here MUST match the order of lines in your data.txt
labels = [
    "Original", 
    "Gemini", 
    "Copilot", 
    "Claude", 
    "ChatGPT", 
    "Perplexity"
]

def load_and_calculate(filename):
    vectors = []
    
    with open(filename, 'r') as f:
        for line in f:
            # Clean up the line: remove brackets/commas and split into numbers
            clean_line = line.replace('[', '').replace(']', '').replace(',', ' ').strip()
            if clean_line:
                # Convert the string of numbers into a numpy array
                vec = np.fromstring(clean_line, sep=' ')
                vectors.append(vec)
    
    if len(vectors) != len(labels):
        print(f"Warning: You have {len(labels)} labels but {len(vectors)} lines in your file!")
    
    # Convert list to a matrix
    matrix = cosine_similarity(vectors)
    
    # We compare everything against the first line ("ORIGINAL")
    original_index = 0
    results = []
    
    for i in range(len(vectors)):
        if i == original_index:
            continue
        score = matrix[original_index][i]
        results.append((labels[i], score))
    
    # Sort results so the best match is at the top
    results.sort(key=lambda x: x[1], reverse=True)
    
    print(f"--- Results: Similarity to {labels[original_index]} ---")
    for label, score in results:
        print(f"{label:15} | Score: {score:.4f}")

# Run the function
load_and_calculate('data.txt')
```

### Euclidean

```python
import numpy as np
import matplotlib.pyplot as plt

labels = [
    "Original", 
    "Gemini", 
    "Copilot", 
    "Claude", 
    "ChatGPT", 
    "Perplexity"
]

# Load embeddings from data.txt
# Assumes each line is space-separated or comma-separated numbers
embeddings = []
with open("data.txt", "r") as f:
    for line in f:
        clean_line = line.strip().replace('[', '').replace(']', '').replace(',', ' ')
        if not clean_line:
            continue
        vec = np.fromstring(clean_line, sep=' ')
        embeddings.append(vec)

embeddings = np.vstack(embeddings)  # shape: (num_vectors, dim)

# Choose the original embedding (first row)
original = embeddings[0]

# Compute Euclidean distances to the original embedding
distances = np.linalg.norm(embeddings - original, axis=1)

print("--- Euclidean distance from Original ---")
for label, dist in zip(labels, distances):
    print(f"{label:12} | {dist:.6f}")

plt.figure(figsize=(9, 4))
plt.bar(labels, distances, color=["tab:blue"] + ["tab:gray"] * (len(labels) - 1))
plt.title("Euclidean Distance from Original")
plt.ylabel("Distance")
plt.xticks(rotation=30, ha="right")
plt.tight_layout()
plt.show()
```
